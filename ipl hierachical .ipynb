```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data = pd.read_csv("auction.csv")
data
```

```python
print("Unique Base prices:", data['Base price'].unique())
print("Unique Winning bids:", data['Winning bid'].unique())
```

```python
X = data[['Base price', 'Winning bid']]
```

```python
Z = linkage(X, method='ward')
```

```python
plt.figure(figsize=(12, 8))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(data['Base price'], data.index, color='skyblue', label='Base price')
plt.scatter(data['Winning bid'], data.index, color='lightgreen', label='Winning bid')
plt.title('Scatter Plot of Base price and Winning bid')
plt.xlabel('Price')
plt.ylabel('Index')
plt.legend()

plt.tight_layout()
plt.show()
```
